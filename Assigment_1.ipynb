```python
import pandas as pd
import numpy as np

# Step 1: Create DataFrame with NA values in Designation
data = {
    "EmpID": [101, 102, 103, 104, 105],
    "Name": ["Amit", "Priya", "Rahul", "Sneha", "Karan"],
    "Designation": ["Data Analyst", np.nan, "ML Engineer", np.nan, "BI Developer"],
    "Exp": [2, 5, 4, 6, 3],
    "Age": [24, 29, 27, 32, 26],
    "Salary": [8000, 15000, 9500, 20000, 7000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Step 2: Check NA values
print("\nMissing Values in Each Column:")
print(df.isna().sum())

# Step 3: Handle NA values in Designation
df["Designation"].fillna("Not Assigned", inplace=True)

print("\nAfter Handling NA in Designation:")
print(df)

# Step 4: Drop rows where Salary > 10000
df = df[df["Salary"] <= 10000]

# Step 5: Reset index
df.reset_index(drop=True, inplace=True)

print("\nFinal DataFrame (Salary <= 10000):")
print(df)
```

```python
import pandas as pd

# Student Details DataFrame
students = pd.DataFrame({
    "StudentID": [1, 2, 3, 4],
    "Name": ["Amit", "Priya", "Rahul", "Sneha"],
    "Age": [20, 21, 19, 22]
})

# Marks Details DataFrame
marks = pd.DataFrame({
    "StudentID": [2, 3, 4, 5],
    "Subject": ["Math", "Science", "English", "History"],
    "Marks": [85, 90, 88, 75]
})

print("Students DataFrame:")
print(students)

print("\nMarks DataFrame:")
print(marks)
```

```python
inner_merge = pd.merge(students, marks, on="StudentID", how="inner")
print("\nInner Merge:")
print(inner_merge)
```

```python
outer_merge = pd.merge(students, marks, on="StudentID", how="outer")
print("\nOuter Merge:")
print(outer_merge)
```

```python
left_merge = pd.merge(students, marks, on="StudentID", how="left")
print("\nLeft Merge:")
print(left_merge)
```

```python
right_merge = pd.merge(students, marks, on="StudentID", how="right")
print("\nRight Merge:")
print(right_merge)
```

```python
concat_rows = pd.concat([students, marks], axis=0)
print("\nRow-wise Concat:")
print(concat_rows)
```

```python
concat_columns = pd.concat([students, marks], axis=1)
print("\nColumn-wise Concat:")
print(concat_columns)
```

```python
import pandas as pd

data = {
    "StudentID": [1, 2, 3],
    "Math": [85, 90, 78],
    "Science": [88, 92, 80],
    "English": [75, 85, 70]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
melted_df = pd.melt(df,
                    id_vars=["StudentID"],
                    var_name="Subject",
                    value_name="Marks")

print("\nMelted DataFrame:")
print(melted_df)
```

```python
data2 = {
    "StudentID": [1, 1, 2, 2, 3, 3],
    "Subject": ["Math", "Science", "Math", "Science", "Math", "Science"],
    "Marks": [85, 88, 90, 92, 78, 80]
}

df2 = pd.DataFrame(data2)
print("Original Long DataFrame:")
print(df2)
```

```python
pivot_df = df2.pivot(index="StudentID",
                     columns="Subject",
                     values="Marks")

print("\nPivoted DataFrame:")
print(pivot_df)
```
